### 解法一

```python
ans=randint(1,101) #從1到100隨機抽取一個數字
print(ans) #可有可無，只是為了驗證程式是否符合預期
num = int(input("請從1～100選一個數字："))

while True: #while迴圈的條件為True
    if num > 100 or num < 1:
        print("阿就叫你從1～100選一個數字，不要亂喔＾＾")
        num=int(input("再輸入一個："))
    elif num > (ans+20):
        print("太大了")
        num=int(input("再輸入一個："))
    elif num > ans and (num-20) <= ans:
        print("有點大，但要接近嘍！")
        num=int(input("再輸入一個："))
    elif num < (ans-20):
        print("太小了") 
        num=int(input("再輸入一個："))
    elif num < ans and (num+20) >= ans:
        print("有點小，但要接近嘍！")
        num=int(input("再輸入一個："))   
    else:
        print("答對嘍～不過也沒有獎金就是了哈哈哈ㄏㄏㄏ嘿嘿嘿")
        break
```

### 解法二

```python
ans=randint(1,101)
print(ans)
num=int(input("請從1～100選一個數字："))

while num != ans: #while迴圈的條件為num!=ans
    if num > 100 or num < 1:
        print("阿就叫你從1～100選一個數字，不要亂喔＾＾")
        num=int(input("再輸入一個："))
    elif num > (ans+20):
        print("太大了")
        num=int(input("再輸入一個："))
    elif num > ans and (num-20) <= ans:
        print("有點大，但要接近嘍！")
        num=int(input("再輸入一個："))
    elif num < (ans-20):
        print("太小了") 
        num=int(input("再輸入一個："))
    elif num < ans and (num+20) >= ans:
        print("有點小，但要接近嘍！")
        num=int(input("再輸入一個："))
print("答對嘍～不過也沒有獎金就是了哈哈哈ㄏㄏㄏ嘿嘿嘿") 
```

### NOTE&PROBLEM :

- 以上兩解的差別在於  
  解法一while迴圈的條件為True，迴圈永遠成立，在Break以前會一直loop，  
  解法二while迴圈的條件為num不等於ans，故只要符合條件就會停止loop。

# 挑戰題二：我有一棵樹

### 解法一

```python
def tree_1(x):
    star=1
    print((x-1)*(" ")+star*"*"+(x-1)*(" "))
    while star<=x-1:
        star+=1
        print((x-star)*(" ")+(2*star-1)*"*"+(x-star)*(" "))
    return 
```

```python
tree_1(6)
```

```python
tree_1(10)
```

### 解法二

```python
def tree_2(x):
    for star in range(1,x+1):
        print((x-star)*(" ")+(2*star-1)*"*"+(x-star)*(" "))  
```

```python
tree_2(6)
```

```python
tree_2(10)
```

### NOTE&PROBLEM :

- 第一解是用while迴圈，  
  第二解是用for迴圈。

# 你的身分證字號正確嗎？

身分證字號的每個字元代表的意義如下： A123456789 

(1) 第一個字元代表地區，轉換方式為： A 轉換成 1,0 兩個字元， B 轉換成 1,1……  



(2) 第二個字元代表性別， 1 代表男性， 2 代表女性    

(3) 第三個字元到第九個字元為流水號碼。  

(4) 第十個字元為檢查號碼。  

檢查碼產生的規則為：  

(1)  A的兩個字元分別＊1和＊9  
     1＊8  
     2＊7  
     3＊6  
     4＊5  
     5＊4  
     6＊3  
     7＊2  
     8＊1  
     9為檢查碼  

(2)權數後之積相加  

(3)相加後之值除以模數 10 取其餘數  

(4)由模數減去餘數得檢查號碼，若餘數為 0 時，則設定其檢查碼為 0  

### 解法一

```python
id=str(input("請輸入你的身分證字號:"))
dict={"A":[1,0] , "B":[1,1] , "C":[1,2] , "D":[1,3] , "E":[1,4] , "F":[1,5] , "G":[1,6] , "H":[1,7] ,
      "I":[3,4] , "J":[1,8] , "K":[1,9] , "L":[2,0] , "M":[2,1] , "N":[2,2] , "O":[3,5] , "P":[2,3] ,
      "Q":[2,4] , "R":[2,5] , "S":[2,6] , "T":[2,7] , "U":[2,8] , "V":[2,9] , "W":[3,2] , "X":[3,0] , 
      "Y":[3,1] , "Z":[3,3]}
#建立字典for第一個字元對照的數字
fir = dict[id[0]]#將輸入的值對照字典
SUM = int(fir[0])*1+int(fir[1])*9+int(id[1])*8+int(id[2])*7+int(id[3])*6+int(id[4])*5+int(id[5])*4+int(id[6])*3+int(id[7])*2+int(id[8])*1
#將流水號等加權後相加
remain = SUM%10 #相加之值取餘數
chk = 10-remain #由模數減去餘數得檢查號碼，若餘數為 0 時，則設定其檢查碼為 0

if int(id[1]) == 1 or int(id[1]) == 2:     #剔除1或2以外的數字出現在第二個字元的id
    if chk == int(id[9]) and chk != 10:    #檢查chk是否符合檢查碼
        print("正確")
    elif chk == 10:                        #在chk出現10的情況下檢查id是否正確
        if int(id[9])==0:
            print("正確")
        else:
            print("錯誤")
    else:
        print("錯誤")
else:
    print("你是在搞笑？")
```
